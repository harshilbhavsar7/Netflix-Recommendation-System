```python
import requests
import pandas as pd

# API key
API_KEY = "e6c711826e918a2e5bfdf2788915bbfa"

# TMDB endpoint for popular movies
url = f"https://api.themoviedb.org/3/movie/popular?api_key={API_KEY}&language=en-US&page=1"

# Fetch data from TMDB API
response = requests.get(url)

# Check if request was successful
if response.status_code == 200:
    # Get the response data in JSON format
    data = response.json()

    # Create a DataFrame to store movie details
    movies_df = pd.DataFrame(columns=["Title", "Overview", "Release Date", "Popularity", "Vote Count", "Vote Average", "Poster Path", "Genres", "Production Companies", "Cast Members"])

    # Extract movie details from the response data and add them to the DataFrame
    for movie in data["results"]:
        title = movie["title"]
        overview = movie["overview"]
        release_date = movie["release_date"]
        popularity = movie["popularity"]
        vote_count = movie["vote_count"]
        vote_average = movie["vote_average"]
        poster_path = movie["poster_path"]

        # Fetch genres for the movie
        genres_url = f"https://api.themoviedb.org/3/movie/{movie['id']}/genres?api_key={API_KEY}"
        genres_response = requests.get(genres_url)
        genres_data = genres_response.json()
        genres_list = [genre["name"] for genre in genres_data["genres"]]

        # Fetch production companies for the movie
        production_companies_url = f"https://api.themoviedb.org/3/movie/{movie['id']}/production_companies?api_key={API_KEY}"
        production_companies_response = requests.get(production_companies_url)
        production_companies_data = production_companies_response.json()
        production_companies_list = [company["name"] for company in production_companies_data]

        # Fetch cast members for the movie
        credits_url = f"https://api.themoviedb.org/3/movie/{movie['id']}/credits?api_key={API_KEY}"
        credits_response = requests.get(credits_url)
        credits_data = credits_response.json()
        cast_list = [cast_member["name"] for cast_member in credits_data["cast"]]

        movies_df = movies_df.append({
            "Title": title,
            "Overview": overview,
            "Release Date": release_date,
            "Popularity": popularity,
            "Vote Count": vote_count,
            "Vote Average": vote_average,
            "Poster Path": poster_path,
            "Genres": genres_list,
            "Production Companies": production_companies_list,
            "Cast Members": cast_list
        }, ignore_index=True)

    # Display the DataFrame
    print(movies_df)
else:
    print("Failed to fetch data from TMDB API.")
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# JustWatch website URL
url = "https://www.justwatch.com/us"

# Search query parameters
query = "netflix"   # Change this to your desired streaming service

# Fetch data from JustWatch website
response = requests.get(f"{url}/providers/{query.lower()}")

# Check if request was successful
if response.status_code == 200:
    # Parse the HTML content using BeautifulSoup
    soup = BeautifulSoup(response.content, "html.parser")

    # Find the div containing movie and TV show results
    search_results = soup.find_all("div", {"class": "title-card"})

    # Create a DataFrame to store movie and TV show details
    movies_df = pd.DataFrame(columns=["Title", "Year", "Age Rating", "IMDb Rating", "Runtime", "Genre"])

    # Extract movie and TV show details from the HTML search results and add them to the DataFrame
    for result in search_results:
        title = result.find("h2", {"class": "title"}).text.strip()
        year = result.find("span", {"class": "metadata"}).text.strip()
        age_rating = result.find("div", {"class": "certification"}).text.strip()
        imdb_rating = result.find("div", {"class": "imdb-rating"}).text.strip()
        runtime = result.find("div", {"class": "runtime"}).text.strip()
        genre = result.find("div", {"class": "genres"}).text.strip()

        movies_df = movies_df.append({
            "Title": title,
            "Year": year,
            "Age Rating": age_rating,
            "IMDb Rating": imdb_rating,
            "Runtime": runtime,
            "Genre": genre
        }, ignore_index=True)

    # Display the DataFrame
    print(movies_df)
else:
    print("Failed to fetch data from JustWatch.")
```

```python
import requests
import json
import pandas as pd

# JustWatch website URL and API endpoint
url = "https://apis.justwatch.com/content"
endpoint = "/titles/en_US/popular"

# Search query parameters
provider_id = "8"  # Netflix provider ID
content_types = "show"
page_size = "20"  # Number of results per page
page = "1"  # Page number

# Make a request to the JustWatch API
response = requests.get(f"{url}{endpoint}?body={{\"page\":{page},\"page_size\":{page_size},\"content_types\":[\"{content_types}\"],\"providers\":[\"{provider_id}\"]}}")

# Check if request was successful
if response.status_code == 200:
    # Parse the JSON response
    response_json = json.loads(response.text)

    # Create a DataFrame to store show details
    shows_df = pd.DataFrame(columns=["Title", "Year", "Age Rating", "IMDb Rating", "Runtime", "Genre", "Cast"])

    # Extract show details from the JSON response and add them to the DataFrame
    for result in response_json["items"]:
        title = result["title"]
        year = result["original_release_year"]
        age_rating = result["age_certification"]
        imdb_rating = result["scoring"]["imdb"]
        runtime = f"{result['runtime']} min" if result["runtime"] else None
        genre = ", ".join(result["genre_ids"])
        cast = result["credits"]

        shows_df = shows_df.append({
            "Title": title,
            "Year": year,
            "Age Rating": age_rating,
            "IMDb Rating": imdb_rating,
            "Runtime": runtime,
            "Genre": genre,
            "Cast": cast
        }, ignore_index=True)

    # Display the DataFrame
    print(shows_df)
else:
    print("Failed to fetch data from JustWatch API.")
```
