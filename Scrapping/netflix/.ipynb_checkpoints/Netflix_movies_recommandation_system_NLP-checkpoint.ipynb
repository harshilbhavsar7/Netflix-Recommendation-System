## Recommendation according to cosine_similarities

```python
# Vectorize the text data using TF-IDF
vectorizer = TfidfVectorizer(stop_words="english")
vectorized_data = vectorizer.fit_transform(netflix_df["text"])
```

```python
vectorized_data
```

```python
# Compute the cosine similarity matrix
cosine_similarities = cosine_similarity(vectorized_data)
```

```python
cosine_similarities
```

```python
# Define the number of topics for LDA
num_topics = 10

# Fit LDA to the vectorized data
lda = LatentDirichletAllocation(n_components=num_topics)
lda.fit(vectorized_data)
```

```python
# Use LDA to predict topic distribution for a new document
new_document = "The guardians of the galaxy were roaming in the space after the fall of the Galactic civilization"
new_vectorized = vectorizer.transform([new_document])
new_topic_distribution = lda.transform(new_vectorized)
print(new_topic_distribution)
```

```python
# Define a function to recommend movies based on a given movie title
def recommend_movies(movie_title, num_recommendations=5):
    # Check if the movie title is in the DataFrame
    if movie_title not in netflix_df["movie_or_tv_show"].values:
        print(f"Error: Movie '{movie_title}' not found in dataset.")
        return

    # Get the index of the movie
    movie_index = netflix_df[netflix_df["movie_or_tv_show"]==movie_title].index[0]

    # Get the cosine similarities for the movie
    cosine_scores = list(enumerate(cosine_similarities[movie_index]))

    # Sort the cosine scores in descending order
    sorted_scores = sorted(cosine_scores, key=lambda x: x[1], reverse=True)

    # Get the indices of the top movies
    top_movie_indices = [i[0] for i in sorted_scores[1:num_recommendations+1]]

    # Get the names of the top movies
    top_movie_names = netflix_df.loc[top_movie_indices]["movie_or_tv_show"].tolist()

    # Print the recommended movies
    print(f"Recommended movies based on '{movie_title}':")
    for i, movie_name in enumerate(top_movie_names):
        print(f"{i+1}. {movie_name}")
```

```python
recommend_movies("Lucifer")
```

## More refined approach to show a table

```python
import pandas as pd

# Define a function to recommend movies based on a given movie title
def recommend_movies_refined(movie_title, num_recommendations=10):
    # Check if the movie title is in the DataFrame
    if movie_title not in netflix_df["movie_or_tv_show"].values:
        print(f"Error: Movie '{movie_title}' not found in dataset.")
        return

    # Get the index of the movie
    movie_index = netflix_df[netflix_df["movie_or_tv_show"]==movie_title].index[0]

    # Get the cosine similarities for the movie
    cosine_scores = list(enumerate(cosine_similarities[movie_index]))

    # Sort the cosine scores in descending order
    sorted_scores = sorted(cosine_scores, key=lambda x: x[1], reverse=True)

    # Get the indices of the top movies
    top_movie_indices = [i[0] for i in sorted_scores[1:num_recommendations+1]]

    # Get the names, vote averages, first air dates, overviews, and similarities of the top movies
    top_movie_data = netflix_df.loc[top_movie_indices, ["movie_or_tv_show", "vote_average", "overview"]]
    top_movie_data["Similarity"] = [i[1] for i in sorted_scores[1:num_recommendations+1]]
    top_movie_data.reset_index(drop=True, inplace=True)

    # Rename the columns for readability
    top_movie_data.columns = ["Name", "Vote Average", "Overview", "Similarity"]

    # Return the top movie data as a DataFrame
    return top_movie_data
```

```python
recommend_movies_refined("Lucifer")
```

```python
import plotly.graph_objects as go
```

```python

def create_table(df):
    fig = go.Figure(data=[go.Table(
        columnorder=[1, 2, 3, 4],
        columnwidth=[20, 20, 30, 30],
        header=dict(values=list(['Name', 'Vote Average', 'Overview', 'Similarity']),
                    line_color='black', font=dict(color='black', family="Gravitas One", size=20), height=40,
                    fill_color='#FF6865',
                    align='center'),
        cells=dict(values=[df["Name"], df["Vote Average"], df["Overview"], df["Similarity"]],
                   font=dict(color='black', family="Lato", size=16),
                   fill_color='#FFB3B2',
                   align='left'))
    ])

    fig.update_layout(height=700,
                      title={'text': "Top 10 Movie Recommendations", 'font': {'size': 22, 'family': 'Gravitas One'}},
                      title_x=0.5
                      )
    fig.show()
```

```python
create_table(recommend_movies_refined("Lucifer"))
```
